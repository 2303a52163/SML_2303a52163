```python
ds=pd.read_csv("/content/Salary_Data.csv")
x= ds['YearsExperience'].values.reshape(-1, 1)
y = ds['Salary'].values.reshape(-1, 1)
```

```python
ds.plot(kind='scatter',title="Relation between YearsExperience and Salary", x='YearsExperience', y='Salary', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
```

#**1.Read the data with pandas and find features and target variables**

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
import matplotlib.pyplot as plt
data = pd.DataFrame({
 'Feature 1': [1.2, 2.4, 3.5, 4.1, 1.9, 3.7, 2.1, 3.3, 2.8, 4.0],
 'Feature 2': [2.3, 1.9, 2.7, 3.0, 2.8, 2.5, 3.2, 2.4, 3.1, 3.3],
 'Feature 3': [3.1, 2.8, 1.5, 3.6, 2.5, 1.9, 2.2, 3.0, 1.8, 2.7],
 'Feature 4': [4.2, 3.5, 2.9, 4.8, 3.2, 4.0, 4.1, 4.5, 3.6, 4.9],
 'Target (Y)': [15.6, 13.1, 12.4, 18.2, 14.1, 16.0, 14.5, 17.3, 14.9, 19.2]
 })
```

# **2.Plot a graph between features and target**

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs[0, 0].scatter(data['Feature 1'], data['Target (Y)'])
axs[0, 0].set_title('Feature 1 vs Target')
axs[0, 1].scatter(data['Feature 2'], data['Target (Y)'])
axs[0, 1].set_title('Feature 2 vs Target')
axs[1, 0].scatter(data['Feature 3'], data['Target (Y)'])
axs[1, 0].set_title('Feature 3 vs Target')
axs[1, 1].scatter(data['Feature 4'], data['Target (Y)'])
axs[1, 1].set_title('Feature 4 vs Target')
fig.tight_layout()
plt.show()
```

# **3.Find Best fit line using linear regression.**

```python
X = data[['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4']]
y = data['Target (Y)']
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Squared Error: {mse:.2f}')
print(f'Mean Absolute Error: {mae:.2f}')
```

# **4.Fond MSE , MAE for test size (20,30)**

```python
test_sizes = [0.2, 0.3]
for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    print(f'Test Size: {test_size:.1f}')
    print(f'Mean Squared Error: {mse:.2f}')
    print(f'Mean Absolute Error: {mae:.2f}')
    print()
```
